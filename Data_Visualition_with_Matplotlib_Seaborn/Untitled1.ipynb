```python
fiyat_btc =np.random.randint(1000,5000,12)
fiyat_eth = np.random.randint(1000,5000,12)
uzunluk =np.arange(fiyat_btc.shape[0])

plt.plot(fiyat_btc,linestyle="--",color="blue",label="BTC")
plt.plot(fiyat_eth,linestyle="--",color="red",label="ETH")
plt.xlabel("Aylar")
plt.ylabel("Fiyat/TL")
plt.title("2024 yılı kripto para fiyatlari")
ay_bilgisi=["ocak","şubat","mart","nisan","mayıs","haziran","temmuz","agustos","eylül","ekim","kasım","aralık"]
plt.xticks(uzunluk,ay_bilgisi,rotation=45)
plt.legend()
plt.show()
```

```python
#Bar(çubuk)Grafikleri -1.gösterim:üst üste yazarak gösterme

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

uzunluk = np.arange(fiyat_btc.shape[0])

plt.bar(uzunluk,fiyat_btc,label="BTC")
plt.bar(uzunluk,fiyat_eth,label="ETH")
plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()
plt.show()
```

```python

fiyat_btc = np.random.randint(100,250,5)
fiyat_eth = np.random.randint(100,250,5)

dizi_isim=["Bitcoin","Ethereum"]

ort_btc=np.mean(fiyat_btc)
ort_eth=np.mean(fiyat_eth)


new_dizi=[ort_btc,ort_eth]

plt.bar(dizi_isim,new_dizi,color=["orange","green"],label=["BTC","ETH"])
#plt.xticks(uzunluk,["2020","2021","2022","2023","2024"])
plt.legend()

for i in range(len(dizi_isim)):
    plt.text(i,new_dizi[1],f"{new_dizi[i]:.2f}",ha="center",va="bottom")
plt.show()
```

```python
#pie Chart(Pasta Grafiği)
#bir dizi halinde degerleri veririz
#bu dizideki degerlerin toplamı %100 oarak kabul edilir
#sonra her bir degerin yüzdesi hesaplanır ve o yüzdeye göre
#bir dilim büyüklüğü belirlenir

giderler=[180,550,350,130,290]
etiketler=["Muhasebe","AR-GE","İnsan Kaynakları","Kalite Yönetimi","Bilişim Teknolojileri (IT)"]

ayırt_etme=[0,0,0,0.2,0 ]

#İngilizce yüzde gösterimi için format=>autopct="%1.1f%%"
#Türkçe yüzde gösterimi için format=>autopct="%%%1f"

plt.pie(giderler,labels=etiketler,autopct="%%%.1f",shadow=True,explode=ayırt_etme)
plt.show()
```
